```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# villes.csv

```python
data =pd.read_csv('villes.csv', sep=";")
```

```python
data
```

```python
#slice data
X = data.iloc[:,1:13]
labels = data.iloc[:, 0]
```

```python
X
```

```python
# Standardizing the data
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA()
```

```python
components = pca.fit_transform(X)
```

```python
#Trouver le nombre daxes  retenir pour conserver un minimum de 90% de linformation reprsente dans le nuage initial.
exp_varaince = pca.explained_variance_ratio_*100
cum_sum_eigenvalues = np.cumsum(exp_varaince)
plt.bar(range(1,len(exp_varaince)+1),exp_varaince, align='center', label='Individual explained variance')
plt.step(range(1,len(cum_sum_eigenvalues)+1), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.title('Explained variance PCA')
plt.xlabel("Component")
plt.ylabel('Explained varaice ratio, %')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#paramtre X=dataset ; labels = nom de l'index; col_varaibles = columns de nom des variables; 
#mindata = minimum de donnes en pourcent

def acp_visual(dataset,labels_col,col_variables,mindata_pourcent) :
    dataset = StandardScaler().fit_transform(dataset)
    pca = PCA()
    components = pca.fit_transform(dataset)
    exp_var =pca.explained_variance_ratio_*100
    cum_sum_eigenvalues = np.cumsum(exp_var)
    #Trouver le nombre daxes  retenir pour conserver un minimum de x% (selon paramtre) de linformation reprsente dans le nuage initial.
    plt.bar(range(1,len(exp_var)+1),exp_var, align='center', label='Individual explained variance')
    plt.step(range(1,len(cum_sum_eigenvalues)+1), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
    plt.title('Explained variance PCA (ACP)')
    plt.xlabel("Component")
    plt.ylabel('Explained varaice ratio, %')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
    #plot une figure de variances ratio
    nb_comp =0
    #trouver nombre de components pricipaux si sum des minimum components depasse ou gal  90% des donnes
    for i in range(len(cum_sum_eigenvalues)) :
        if cum_sum_eigenvalues[i]>=mindata_pourcent :
            nb_comp=i+1
            break
    pca = PCA(n_components=nb_comp)
    components = pca.fit_transform(dataset)
    #variables_loading sur les components principaux
    loadings = pca.components_
    num_pc = pca.n_features_
    pc_list = ["CP"+str(i) for i in list(range(1, num_pc+1))]
    loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
    loadings_df['variable'] = col_variables.values
    loadings_df = loadings_df.set_index('variable')
    print(loadings_df)
    
    if nb_comp==2 :
        fig = plt.figure(figsize=(8,6))
        ax= fig.add_subplot(1,1,1)
        ax.grid()
        ax.set_xlabel('Component principal 1 ('+str(round(exp_var[0],2))+'%)')
        ax.set_ylabel('Component principal 2 ('+str(round(exp_var[1],2))+'%)')
        ax.set_title("2 Components ACP")
        """
        plt.figure(figsize = (10,9))
        plt.title("2 Components ACP")
        plt.xlabel('Component principal 1 ('+str(round(exp_var[0],2))+'%)')
        plt.ylabel('Component principal 2 ('+str(round(exp_var[1],2))+'%)')
        """
        plt.scatter(components[:, 0], components[:, 1])
        for l, x, y in zip(labels_col,components[:, 0], components[:, 1]):
            plt.annotate(l, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
        plt.show()
        #plot variables correlation avec chaque components pricipaux
        plt.rcParams["figure.figsize"] = [8, 6]
        plt.rcParams["figure.autolayout"] = True
        fig, ax = plt.subplots()
        loadings_df.plot.scatter('CP1', 'CP2', ax=ax, grid=True)
        for k, v in loadings_df.iterrows():
            ax.annotate(k, v)
        plt.title('Variables corrlations sur les components principaux')
        plt.show()
    elif nb_comp==3 :
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111, projection='3d')
        for i in range(len(components)): #plot each point + it's index as text above
            ax.scatter(components[i,0],components[i,1],components[i,2],color='b') 
            ax.text(components[i,0],components[i,1],components[i,2],  '%s' % (str(labels_col[i])), size=8, zorder=1,  
            color='k') 
        ax.set_title('3 Components ACP')
        ax.set_xlabel("CP1 ("+str(round(exp_var[0],2))+"%)")
        ax.set_ylabel("CP2 ("+str(round(exp_var[1],2))+"%)")
        ax.set_zlabel("CP3 ("+str(round(exp_var[2],2))+"%)")
        plt.show()
        #plot variables correlation avec chaque components pricipaux
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111, projection='3d')
        for k, v in loadings_df.iterrows():
            ax.scatter(v[0],v[1],v[2],color='b') 
            ax.text(v[0],v[1],v[2], '%s' %(str(k)), size=8, zorder=1,  
            color='k') 
        ax.set_title('Variables corrlations sur les components principaux')
        ax.set_xlabel("CP1 ("+str(round(exp_var[0],2))+"%)")
        ax.set_ylabel("CP2 ("+str(round(exp_var[1],2))+"%)")
        ax.set_zlabel("CP3 ("+str(round(exp_var[2],2))+"%)")
        plt.show()
        
    return loadings_df
    
```
